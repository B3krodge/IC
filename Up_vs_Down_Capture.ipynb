```python
import polygon
import requests
import csv
import pandas as pd
import numpy as np
```

```python

def get_data(from_date, to_date, ticker):

# build the API endpoint URL with the specified parameters
  url = f'https://api.polygon.io/v2/aggs/ticker/{ticker}/range/1/day/{from_date}/{to_date}?apiKey={POLYGON_KEY}'

  response = requests.get(url)
  data = response.json()

# extract the daily stock prices from the API response
  daily_prices = data['results']
  # print (daily_prices)
#Create file
  file_name = (ticker + '.csv')
  # print (file_name)
# write the daily prices to a CSV file
  with open(file_name, mode='w', newline='') as file:
      writer = csv.writer(file)
      writer.writerow(['Date', 'Closing Price'])
      for price in daily_prices:
          date = price['t']
          closing_price = price['c']
          writer.writerow([date, closing_price])
  df = pd.read_csv(file_name)

# print the DataFrame to the console
  return(df)
```

```python
### parameters:
### from_date = start of investment
### to_date = end of investment
### ticker = ticker you are trying to examine
def compare(from_date, to_date, ticker):
  closing_prices1 = np.array(get_data(from_date, to_date, 'SPY').loc[:, 'Closing Price'])
  closing_prices2 = np.array(get_data(from_date, to_date, ticker).loc[:, 'Closing Price'])
  up = 0
  upCount = 0
  down = 0
  downCount = 0
  capture = 0
  for i in range(len(closing_prices1)):
    if (i != 0):
      ##UP
      if (closing_prices1[i] > closing_prices1[i-1]):
        percent1 = (closing_prices1[i] - closing_prices1[i-1])/ closing_prices1[i-1]
        percent2 = (closing_prices2[i] - closing_prices2[i-1])/ closing_prices2[i-1]
        up += (percent2/percent1)
        upCount += 1
      ##DOWN
      elif (closing_prices1[i] < closing_prices1[i-1]):
        percent1 = (closing_prices1[i] - closing_prices1[i-1])/ closing_prices1[i-1]
        percent2 = (closing_prices2[i] - closing_prices2[i-1])/ closing_prices2[i-1]
        down += (percent2/percent1)
        downCount += 1
  if (up != 0 and down != 0):
      up = up/upCount
      down = down/downCount
      capture = up/down
      capture *= 100
  return "Up vs Down capture = " + str(capture)
```
